### Padding 
  By applying the convolution operation without padding we loose the information at the corner of the images.The following diagram show that without padding the corner pixels are covered only 1 times, edge pixel are covered 2 times and all the remaining pixels are covered 3 times.
  <img src="images/No_padding.png">
 
 By padding we ensure that all the pixels are covered equally as shown in the figure below
   <img src="images/With_padding.png">

  Padding basically extends the area of an image in which a convolutional neural network processes. We require padding because of he following reasons:
  1. It help us in preserving the inforation at the corner of the image. Without padding, the information at the corner of the image will be lost.
  2. Without padding, the height $n_H$ and width $n_W$ decreases after the convolution operation. By using the padding, we can apply the convolution operation without decreasing the height and width of the image.  
  
 


### Pooling
   The pooling layer helps in reducing the height and width of the image. Thus, it helps in reducing the number of parameters and also make invariant to the position in he input.
   <br>
   <b> The pooling layer has no parameter </b> .There are two kinds of pooling :
   1. Max pooling : The filter of size $(f,f)$ is slided over the input with a stride and we take the maximum in that region.
   
   2. Average pooling : The filter of size $(f,f)$ is slided over the input with a sride and we take the average in that region.

  <b>Shape of the output:</b>
  <br>
    Consider that we have input image of dimension $(n_H,n_W,n_C)$ and filter of size (f,f) with a stride of s. Then the output will have a dimesion of $(n_H',n_W',n_C)$ where $n_H$ and $n_W$ are computed as follows :
   $$ n_H' = \lfloor \frac{n_H - f }{s} \rfloor +1 $$
   $$ n_W' = \lfloor \frac{n_W - f }{s} \rfloor +1 $$
    
    
 <b>Application of Pooling:</b><br>
    1. Average pooling is used for feature extraction while going from Conv layers to Fully connected layers.<br>
    2. Max pooling is used in between the Conv layers.

### Convolution for 3D image
  In the above section, we have presented the convolution for a 2D image. But images are 3D in shape. Here we would show the convolution operation for 3D images. With the help of convolution, we can increase the depth of image and reduce the height and width.


<img src="images/conv_nn.png" style="width:350px;height:200px;">

  For the case of 3D image,  the input image of shape $(n_H,n_W,n_C)$ is convolved with a filter of shape $(f,f,n_C)$ to give an output of shape $(n_H',n_W',n_C') $.
  
  <br>
  <b>Shape of the output:</b>
  <br>
  The $n_H', n_W', n_C'$  coressponds to the output height, output width and output channels. They are computed as follows:
   $$ n_H' = \lfloor \frac{n_H - f + 2 \times pad}{s} \rfloor +1 $$
   $$ n_W' = \lfloor \frac{n_W - f + 2 \times pad}{s} \rfloor +1 $$
   $$ n_C' = \text{number of filters used in the convolution}$$
 
 
  <b> No of parameters: </b>
     <br>
     The no of parameters is equal to the dimension of weights + bias of the filters.<br> 
     For each filter, weight of filter is having shape of $f*f*n_C$ and bias shape is 1.
     As there are $n_C'$ no of filter in the output so the total number of parameter is :
       $(f*f*n_C +1)*n_C'$
